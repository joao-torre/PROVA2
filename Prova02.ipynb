Notei que saída não está ordenada corretamente em vários casos
Em alguns casos aparecem números repetidos ou valores sumindo

```python
# erro 1 min_idx = i + 1 (não considerar a posição i)
def selection_sort_corrigido_1(arr):
    for i in range(len(arr)-1):
        min_idx = i    # CORREÇÃO: começar em i
        for j in range(i, len(arr)):
            if arr[j] < arr[1]:
                min_idx = j
        arr[min_idx] = arr[i]
        arr[i] = arr[min_idx]
    return arr

for t in [[3,1,4,2],[5,4,3,2,1]]:
    print('entrada:', t, '-> saída:', selection_sort_corrigido_1(t.copy()))
```

```python
#erro 2 comparação fixa arr[1] (usar arr[min_idx])
def selection_sort_corrigido_2(arr):
    for i in range(len(arr)-1):
        min_idx = i
        for j in range(i, len(arr)):
            if arr[j] < arr[min_idx]:  # CORREÇÃO: comparar com arr[min_idx]
                min_idx = j
        arr[min_idx] = arr[i]
        arr[i] = arr[min_idx]
    return arr

for t in [[3,1,4,2],[5,4,3,2,1]]:
    print('entrada:', t, '-> saída:', selection_sort_corrigido_2(t.copy()))
```

```python
#erro 3 swap (troca) errado
def selection_sort_corrigido_3(arr):
    for i in range(len(arr)-1):
        min_idx = i
        for j in range(i, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # TROCA CORRETA:
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

for t in [[3,1,4,2],[5,4,3,2,1],[2,2,2]]:
    print('entrada:', t, '-> saída:', selection_sort_corrigido_3(t.copy()))
```

```python
#erro 4 laço interno começar em i+1 (O laço interno costuma começar em i+1)
def selection_sort_final(arr):
    n = len(arr)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):  # começa no próximo elemento
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
    
# Testes finais
testes = [[], [1], [2,1], [3,1,4,2], [5,4,3,2,1], [2,2,2], [9,-1,5,0,3,3]]

for t in testes:
    print(selection_sort_final(t.copy()))

```
